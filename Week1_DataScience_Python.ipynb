```

import pandas as pd
import matplotlib.pyplot as plt
from statistics import mode
```

```
df = pd.read_csv('/mnt/data/student_marks.csv')
df
```

```

# Descriptive Statistics
mean_scores = df[['Math','Science','English']].mean()
median_scores = df[['Math','Science','English']].median()
mode_scores = df[['Math','Science','English']].mode().iloc[0]
std_scores = df[['Math','Science','English']].std()

print("Mean Scores:\n", mean_scores)
print("\nMedian Scores:\n", median_scores)
print("\nMode Scores:\n", mode_scores)
print("\nStandard Deviation:\n", std_scores)
```

```

# Visualizations
df[['Math','Science','English']].hist(bins=5, figsize=(10,6), color='skyblue', edgecolor='black')
plt.suptitle('Score Distributions of Students', fontsize=16)
plt.show()
```

```

# Insights / Summary
print("""
Insights from the dataset:

1. Math scores: Mean ~ {0:.2f}, Std Dev ~ {1:.2f}. Scores are fairly spread with some lower outliers.
2. Science scores: Mean ~ {2:.2f}, Std Dev ~ {3:.2f}. Distribution is slightly higher, few students scoring very high.
3. English scores: Mean ~ {4:.2f}, Std Dev ~ {5:.2f}. Most students perform consistently with a few top scorers.
4. Overall, the students show good performance with some variations. Outliers include low scoring students in Math and Science.
""".format(mean_scores['Math'], std_scores['Math'],
           mean_scores['Science'], std_scores['Science'],
           mean_scores['English'], std_scores['English']))
```
